## Correlations

[Explore correlations between features and the target variable, as well as among features themselves.]


```python
# Example: Plotting a heatmap to show feature correlations
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True)
plt.show()
```

```python
# Define target variable and input features
target = 'delta'
features = [col for col in df.columns if col != target]

# Compute correlations with "delta"
correlations = df.corr(numeric_only=True)[target].drop(target)

# Convert to DataFrame and sort
correlation_df = correlations.to_frame(name='Correlation Coefficient').sort_values(by='Correlation Coefficient', ascending=False)

# Create bar plot
plt.figure(figsize=(8, 5))
sns.barplot(x=correlation_df['Correlation Coefficient'], y=correlation_df.index, palette='coolwarm')
plt.title('Influence of Other Variables on "delta" (Correlation Analysis)')
plt.xlabel('Correlation with delta')
plt.ylabel('Variable')
plt.grid(True)
plt.tight_layout()
plt.show()

correlation_df.round(3)
```
