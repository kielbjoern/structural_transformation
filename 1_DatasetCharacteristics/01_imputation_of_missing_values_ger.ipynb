```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
import pandas as pd
from tensorflow.keras.layers import Layer


# Load the data of Germany.
df_mv = pd.read_csv('germany.csv', sep=";", header=0)

# erase any non-used data in the DataFrame
df_mv = df_mv.drop(df_mv.columns[0:7], axis=1)

print(df_mv.head())
print(df_mv.dtypes)

# List of years to be processed
years = [1882, 1895, 1907, 1925, 1933, 1939]


def process_year(year):
    # Keep only the values where year = variable year.
    df_mv_2 = df_mv[df_mv["year_of_census"] == year].copy()
    df_mv_2.iloc[:, 3:6] = round(df_mv_2.iloc[:, 3:6] / 100, 3)  # convert to percentage

    # I use a Self-Supervised Imputation - Bidirectional RNN

    # === Input data ===
    df_i = df_mv_2.copy()
    df_i = df_i.reset_index()  # Make 'year_of_birth' a column
    df_i = df_i[['year_of_birth', '1. Sector', '2. Sector', '3. Sector']]

    # === Add years ===
    year_range = np.arange(df_i['year_of_birth'].min(), df_i['year_of_birth'].max() + 1)
    df_full = pd.DataFrame({'year_of_birth': year_range})
    df_merged = pd.merge(df_full, df_i, on='year_of_birth', how='left')

    # === Normalized time ===
    df_merged['year_norm'] = (df_merged['year_of_birth'] - df_merged['year_of_birth'].min()) / \
                            (df_merged['year_of_birth'].max() - df_merged['year_of_birth'].min())

    features = df_merged[['1. Sector', '2. Sector', '3. Sector']].values.astype(np.float32)
    mask = ~np.isnan(features)
    values = np.nan_to_num(features, nan=0.0)  # missing values = 0 
    year_feature = df_merged['year_norm'].values[:, np.newaxis].astype(np.float32)

    # === Inputs & target formats ===
    X_input = year_feature[np.newaxis, :, :]            # shape (1, T, 1)
    Y_target = values[np.newaxis, :, :]                 # shape (1, T, 3)
    mask_tensor = mask[np.newaxis, :, :]                # shape (1, T, 3)

    # === Custom loss function ===
    def masked_mse(y_true, y_pred):
        mask = tf.cast(tf.math.not_equal(y_true, 0.0), tf.float32)
        loss = tf.reduce_sum(tf.square(y_true - y_pred) * mask) / tf.reduce_sum(mask)
        return loss

    class TemperatureSoftmax(Layer):
        def __init__(self, temperature=2.0, **kwargs):
            super().__init__(**kwargs)
            self.temperature = temperature

        def call(self, logits):
            return tf.nn.softmax(logits / self.temperature)


    model = tf.keras.Sequential([
        tf.keras.layers.Input(shape=(X_input.shape[1], X_input.shape[2])),  # (T, 1)
        tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(32, return_sequences=True)),
        tf.keras.layers.Dense(3),  # without activation
        TemperatureSoftmax(temperature=2.0)  # smoother Softmax
    ])

    model.compile(optimizer='adam', loss=masked_mse)

    # === Training only on this one time series ===
    model.fit(X_input, Y_target, epochs=500, verbose=1)

    # === Prediction ===
    Y_pred = model.predict(X_input)[0]  # shape: (T, 3)

    # === Imputation: replace missing values ===
    values_imputed = features.copy()
    values_imputed[~mask] = Y_pred[~mask]

    # === Return as DataFrame ===
    df_result = df_merged[['year_of_birth']].copy()
    df_result[['year']] = year   
    df_result[['1. Sector', '2. Sector', '3. Sector']] = values_imputed
    file_name = f"ger_imputed_{year}.csv"
    df_result.to_csv(file_name, index=False)

    print(df_result.head(10))

    import matplotlib.pyplot as plt

    years = df_i.iloc[:, 0].values.astype(np.float32)
    sectors = df_i.iloc[:, 1:4].values.astype(np.float32)

    # Plot the imputed sector shares    
    plt.figure(figsize=(12, 6))
    plt.plot(df_result['year_of_birth'], df_result['1. Sector'], label='1. Sector')
    plt.plot(df_result['year_of_birth'], df_result['2. Sector'], label='2. Sector')
    plt.plot(df_result['year_of_birth'], df_result['3. Sector'], label='3. Sector')
    plt.scatter(years, sectors[:, 0], c='tab:blue', marker='o')
    plt.scatter(years, sectors[:, 1], c='tab:orange', marker='o')
    plt.scatter(years, sectors[:, 2], c='tab:green', marker='o')
    plt.title("Imputed sector shares across cohorts")
    plt.xlabel("Year of birth")
    plt.ylabel("Sector share")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()


# Loop over all years
for year in years:
    process_year(year)
```
